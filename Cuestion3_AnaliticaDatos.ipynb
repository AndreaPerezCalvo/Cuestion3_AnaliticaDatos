2. Primero Importamos la librería pandas, la cual está especializada en el manejo de datos.

5. Ahora realizaremos un gráfico interactivo, en el que se podrá agrandar subir y bajar para ver los datos.
6. Para ello tenemos que instalar en Anaconda la siguiente biblioteca: "altair" tras instalarla(se podrá hacer desde comandos o desde Anaconda Navigator) la importaremos.
7. Una vez importadas tanto altair como pandas, leeremos nuestro archivo csv
8. Creamos la variable chart que llevará el gráfico del csv
9. Vamos a indicarle que serán puntos mediante mark_point y lo codifica con encode.
10. Después indicamos que datos aparecerán en cada eje.
11. Por último haremos nuestro gráfico interactivo con el método interactive()

```python
import altair as alt
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/AndreaPerezCalvo/Cuestion3_AnaliticaDatos/main/datos.csv")
chart = alt.Chart(df)
alt.Chart(df).mark_point().encode(
    x='Make',
    y='Popularity'
).interactive()
```
